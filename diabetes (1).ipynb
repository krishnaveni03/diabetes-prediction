```python
import gradio as gr
import pandas as pd
import numpy as np
```

```python
#reading the csv file
data=pd.read_csv('diabetes.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
#splitting x and y
X=data.iloc[:,:-1]
X
```

```python
Y=data.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

# SVM Algoriithm

```python
from sklearn.svm import SVC
classifier1 = SVC(kernel='rbf')
classifier1.fit(X_train,Y_train)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier1,X=X_train,y=Y_train,cv=10)
print("Accuracy : {:.2f}%".format(accuracies.mean()*100))
print("Standard Deviation : {:.2f}".format(accuracies.std()*100))
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'C': [0.25,0.50,0.75,1.0],'kernel':['rbf','linear']}]
grid_search = GridSearchCV(estimator=classifier1,param_grid=parameters,scoring='accuracy',cv=10,n_jobs=-1)
grid_search.fit(X_train,Y_train)
best_accuracy1 = grid_search.best_score_
best_parameters1 = grid_search.best_params_
print("Best Accuracy : {:.2f}%".format(best_accuracy1*100))
print("Best Parameters : ",best_parameters1)
```

```python
Y_pred1 = classifier1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_pred1))
```

```python
def classify(num):
    if num==0:
        return "No Diabetes"
    if num==1:
        return "Yes,Diabetes Conformed"
```

```python
def predict_diabetes(Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age):
  in_array=np.array([[Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age]])
  predict=classifier.predict(in_array)
  output=classify(predict[0])
  if output=="No Diabetes" :
    return [(0,output)]
  else:
    return [(1,output)]
```

```python
Pregnancies= gr.inputs.Slider(minimum=0,maximum=15,default=0,label="Pregnancies")
Glucose=gr.inputs.Slider(minimum=0,maximum=200,default=99,label="Glucose")
BloodPressure=gr.inputs.Slider(minimum=0,maximum=200,default=100,label="Blood pressure")
SkinThickness=gr.inputs.Slider(minimum=0,maximum=100,default=10,label="Skin Thickness")
Insulin=gr.inputs.Slider(minimum=0,maximum=500,default=0,label="Insulin")
BMI=gr.inputs.Slider(minimum=0,maximum=100,default=1,label="BMI(BODY MASS INDEX)")
Age=gr.inputs.Slider(minimum=0,maximum=100,default=20,label="Age")
DiabetesPedigreeFunction=gr.inputs.Slider(minimum=0,maximum=3,default=0.3,label="Diabetes Pedigree Function ")
```

```python
outcome=gr.outputs.HighlightedText(color_map={"Yes,Diabetes is conformed" : "red",
                                         "No Diabetes" : "lightgreen"})
```

```python
gr.Interface(predict_diabetes,inputs=[Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age],outputs=outcome,live=True).launch(debug=True)
```
